```python
FBchipmini = FBchipcount[~FBchipcount.index.str.contains('Bulk')].copy()
FBchipmini.drop(['FranklinBluffs_40cm_Bulk','FranklinBluffs_55cm_Bulk'],axis=1,inplace=True)
```

```python
import time
threshold = 3000
chipminithres = FBchipmini[FBchipmini>threshold]
chip2 = chipminithres.divide(chipminithres).fillna(0)
c1 = 'SuperContig_NODE_3114_length_7886_cov_11.2671'
c2 = 'SuperContig_NODE_6017_length_5226_cov_107.278'

def dist_calc_chip(df,c1,c2):
    tt = np.sum(df.loc[[c1,c2]].sum()==2)
    nn = np.sum(df.loc[[c1,c2]].sum()==0)
    tn = np.sum(df.loc[c1]-df.loc[c2]==1)
    nt = np.sum(df.loc[c1]-df.loc[c2]==-1)
    
    t1 = tt+tn;
    n1 = nt+nn;
    t2 = tt+nt;
    n2 = tn+nn;
    tot = len(df)
    """num = np.math.factorial(a+b)*np.math.factorial(c+d)*np.math.factorial(a+c)*np.math.factorial(d+b)
    denom = np.math.factorial(a)*np.math.factorial(b)*np.math.factorial(c)*np.math.factorial(d)*np.math.factorial(a+b+c+d)
    p = num/denom"""
    
    
    p = np.math.factorial(n1)/np.math.factorial(tot)*np.math.factorial(n2)/np.math.factorial(nn)/np.math.factorial(nt)*np.math.factorial(t1)/np.math.factorial(tn)*np.math.factorial(t2)/np.math.factorial(tt)
    return p

def get_dist_matrix(df):
    start = time.time()
    dist_matrix = pd.DataFrame(index=df.index,columns=df.index)
    for i,j in itertools.combinations(df.index,2):
        dist = dist_calc_chip(chip2,i,j)
        dist_matrix.loc[i,j],dist_matrix.loc[j,i] = dist,dist
    end = time.time()
    print('calculating distance matrix of '+str(len(df))+' contigs took {:.2f} seconds'.format(end - start))
    return dist_matrix

chip_part = chip2[0:20]
```
