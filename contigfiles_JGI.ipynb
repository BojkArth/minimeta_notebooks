```python
### column_num options:
# 0 = Scaffold ID
# 1 = Scaffold Name
# 2 = Contig origin
# 3 = Genome
# 4 = Gene Count
# 5 = Sequence Length
# 6 = GC Content
# 7 = Read Depth
# 8 = Lineage Domain
# 9 = Lineage Phylum
# 10 = Lineage Class
# 11 = Lineage Order
# 12 = Lineage Family
# 13 = Lineage Genus
# 14 = Lineage Species
# 15 = Lineage Percentage
def getparams(column_num,dataset):
    int_cols = [1,4,5,7]
    float_cols = [6,15]
    output = []
    for i in range(1,len(dataset)):
        output.append(dataset[i][column_num])
    if column_num in float_cols:
        output = np.array([float(i) for i in output])
    if column_num in int_cols:
        output = np.array([int(i) for i in output])
    return(output)
    


```

```python
def separatedomainContigs(inputdata):
    # get the domains
    domains_all = getparams(8,inputdata)
    # get the contig lengths
    contigs_all = getparams(5,inputdata)
    genecount_all = getparams(4,inputdata)
    GCcont_all = getparams(6,inputdata)
    Origin_all = getparams(2,inputdata)
    
    # get the unique domains 
    domaintypes = []
    for f in sorted(set(domains_all)):
        domaintypes.append(f)
        
    # make domain dictionary    
    domdct = {};domgenec = {};domGC = {};domOrigin = {}
    for i in domaintypes:
        domdct['%s' % i] = []
        domgenec['%s' % i] = []
        domGC['%s' % i] = []
        domOrigin['%s' % i] = []
    
    # fill it with the appropriate shit
    for i in range(0,len(domains_all)):
        for j in domaintypes:
            if domains_all[i] == j:
                domdct.setdefault(j, []).append(contigs_all[i])
                domgenec.setdefault(j, []).append(genecount_all[i])
                domGC.setdefault(j, []).append(GCcont_all[i])
                domOrigin.setdefault(j, []).append(Origin_all[i])

    return(domdct,domgenec,domGC,domOrigin)
        
def separateOriginContigs(inputdata):
    # get the domains
    Origin_all = getparams(2,inputdata)
    # get the contig lengths
    contigs_all = getparams(5,inputdata)
    genecount_all = getparams(4,inputdata)
    GCcont_all = getparams(6,inputdata)
    domains_all = getparams(8,inputdata)
    phyla_all = getparams(9,inputdata)
    
    
    # get the unique domains 
    origintypes = []
    for f in sorted(set(Origin_all)):
        origintypes.append(f)
        
    # make domain dictionary    
    orlen = {};orgenec = {};orGC = {};ordom = {};orphyl = {}
    for i in origintypes:
        orlen['%s' % i] = []
        orgenec['%s' % i] = []
        orGC['%s' % i] = []
        ordom['%s' % i] = []
        orphyl['%s' % i] = []
    
    # fill it with the appropriate shit
    for i in range(0,len(domains_all)):
        for j in origintypes:
            if Origin_all[i] == j:
                orlen.setdefault(j, []).append(contigs_all[i])
                orgenec.setdefault(j, []).append(genecount_all[i])
                orGC.setdefault(j, []).append(GCcont_all[i])
                ordom.setdefault(j, []).append(domains_all[i])
                orphyl.setdefault(j, []).append(phyla_all[i])

    return(orlen,orgenec,orGC,ordom,orphyl)
```
